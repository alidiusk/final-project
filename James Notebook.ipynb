## Introduction
I picked public art as my metric for finding the "best" neighborhood in Pittsburgh, because there is a lot around outside and not just in museums. I decided to first measure the best neighborhood based off of quantity of pieces.

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
#Importing the data set from the WPRDC website:
art_per_neighborhood = pd.read_csv("https://data.wprdc.org/datastore/dump/00d74e83-8a23-486e-841b-286e1332a151", index_col = "id")
```

## Step 1: Just the Numbers
After importing the data and making a graph based off of quantity of pieces, the Central Business District was the clear victor. However, I wanted to see if the other neighborhoods would have a chance if a changed what counted as "best."

```python
art_per_neighborhood["neighborhood"].value_counts().sort_index().plot(kind='bar', rot= 'vertical', figsize = (30, 10))
```
